```python
import os
import time
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.amp import GradScaler, autocast
from torchvision import transforms, models
from torchvision.utils import make_grid
from PIL import Image
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import gc
from datetime import datetime

class CelebAHQDataset(Dataset):
    def __init__(self, data_dir, attr_file, transform=None, split='train'):
        self.data_dir = data_dir
        print(f"Data directory: {self.data_dir}")
        self.transform = transform

        print(f"Attribute file: {attr_file}")
        if attr_file.endswith('.xlsx'):
            self.attributes_df = pd.read_excel(attr_file, engine='openpyxl')
        else:
            self.attributes_df = pd.read_csv(attr_file)

        self.attr_names = self.attributes_df.columns[1:].tolist()
        self.attr_filenames = [os.path.basename(str(f)).strip().lower() for f in self.attributes_df.iloc[:, 0].tolist()]
        img_files = [f.strip().lower() for f in os.listdir(self.data_dir) if f.lower().endswith(('.jpg', '.png'))]

        print("\U0001F4BE First 5 filenames from CSV:", self.attr_filenames[:5])
        print("\U0001F5BC First 5 filenames from folder:", img_files[:5])

        filename_to_idx = {name: idx for idx, name in enumerate(self.attr_filenames)}
        self.matched_files, self.matched_indices = [], []

        for img_file in img_files:
            if img_file in filename_to_idx:
                self.matched_files.append(img_file)
                self.matched_indices.append(filename_to_idx[img_file])

        self.attributes = self.attributes_df.iloc[self.matched_indices, 1:]
        split_idx = int(0.8 * len(self.matched_files))
        if split == 'train':
            self.matched_files = self.matched_files[:split_idx]
            self.attributes = self.attributes.iloc[:split_idx]
        else:
            self.matched_files = self.matched_files[split_idx:]
            self.attributes = self.attributes.iloc[split_idx:]

        print("Sample attributes (first 5 samples):", self.attributes.head().values)
        print(f"\u2705 Matched {len(self.matched_files)} images out of {len(img_files)}")

    def __len__(self):
        return len(self.matched_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.data_dir, self.matched_files[idx])
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        attrs = (self.attributes.iloc[idx].values + 1) // 2
        return image, torch.tensor(attrs, dtype=torch.float32)

class EnhancedGenerator(nn.Module):
    def __init__(self, noise_dim=100, cond_dim=40, out_channels=3):
        super().__init__()
        self.main = nn.Sequential(
            nn.Linear(noise_dim + cond_dim, 512 * 8 * 8),
            nn.BatchNorm1d(512 * 8 * 8),
            nn.ReLU(True),
            nn.Unflatten(1, (512, 8, 8)),
            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),  # 8 -> 16
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),  # 16 -> 32
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),   # 32 -> 64
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 32, 4, 2, 1, bias=False),    # 64 -> 128
            nn.BatchNorm2d(32),
            nn.ReLU(True),
            nn.ConvTranspose2d(32, out_channels, 4, 2, 1, bias=False),  # 128 -> 256
            nn.Tanh()
        )
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, (nn.ConvTranspose2d, nn.Linear)):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, noise, cond):
        x = torch.cat([noise, cond], dim=1)
        return self.main(x)

class Discriminator(nn.Module):
    def __init__(self, in_channels=3, cond_dim=40):
        super().__init__()
        self.cond_dim = cond_dim
        self.main = nn.Sequential(
            nn.Conv2d(in_channels + cond_dim, 32, 4, 2, 1, bias=False),   # 256 -> 128
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(32, 64, 4, 2, 1, bias=False),                        # 128 -> 64
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, 4, 2, 1, bias=False),                       # 64 -> 32
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(128, 256, 4, 2, 1, bias=False),                      # 32 -> 16
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(256, 512, 4, 2, 1, bias=False),                      # 16 -> 8
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(512, 512, 4, 2, 1, bias=False),                      # 8 -> 4
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.validity = nn.Conv2d(512, 1, 4, 1, 0, bias=False)

        # Infer flattened features
        with torch.no_grad():
            dummy_input = torch.randn(1, in_channels + cond_dim, 256, 256)
            dummy_out = self.main(dummy_input)
            self.flattened_feat_size = dummy_out.view(1, -1).shape[1]

        self.aux = nn.Sequential(
            nn.Flatten(),
            nn.Linear(self.flattened_feat_size, cond_dim)
        )

        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, (nn.Conv2d, nn.Linear)):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, img, cond):
        cond_exp = cond.view(cond.size(0), cond.size(1), 1, 1).expand(-1, -1, img.size(2), img.size(3))
        x = torch.cat([img, cond_exp], dim=1)
        x = self.main(x)
        validity = self.validity(x)
        aux = self.aux(x)
        return validity, aux

def compute_gradient_penalty(discriminator, real_images, fake_images, cond, device):
    batch_size = real_images.size(0)
    alpha = torch.rand(batch_size, 1, 1, 1, device=device)
    interpolates = alpha * real_images + (1 - alpha) * fake_images
    interpolates = interpolates.requires_grad_(True)
    d_interpolates, _ = discriminator(interpolates, cond)
    gradients = torch.autograd.grad(outputs=d_interpolates, inputs=interpolates,
                                  grad_outputs=torch.ones_like(d_interpolates),
                                  create_graph=True, retain_graph=True)[0]
    gradients = gradients.view(batch_size, -1)
    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()  # Unchanged base penalty
    return gradient_penalty / batch_size  # Normalize by batch size

def train_and_evaluate():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    data_dir = r"C:\Users\USER\Conditional_Face_Generation_CelebA\img_align_celeba\img_align_celeba"
    attr_file = r"C:\Users\USER\Conditional_Face_Generation_CelebA\img_align_celeba\list_attr_celeba.csv"
    batch_size = 128
    num_epochs = 300
    lr = 2e-4  # Reduced learning rate
    noise_dim = 100
    checkpoint_interval = 10
    checkpoint_dir = "gan_checkpoints"
    os.makedirs(checkpoint_dir, exist_ok=True)
    sample_dir = os.path.join(checkpoint_dir, "samples")
    os.makedirs(sample_dir, exist_ok=True)

    transform = transforms.Compose([
        transforms.Resize((256, 256)),  # Resize 178x218 -> 256x256
        transforms.ToTensor(),
        transforms.Normalize([0.5]*3, [0.5]*3)
    ])

    train_dataset = CelebAHQDataset(data_dir, attr_file, transform, split='train')
    val_dataset = CelebAHQDataset(data_dir, attr_file, transform, split='val')
    print(f"Train dataset size: {len(train_dataset)}, Val dataset size: {len(val_dataset)}")
    
    train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
    for images, attrs in train_loader:
        plt.imshow(make_grid(images).permute(1, 2, 0).cpu().numpy())
        plt.title([train_dataset.attr_names[i] for i in attrs[0].nonzero()[0]])
        plt.show()
        break

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    generator = EnhancedGenerator().to(device)
    discriminator = Discriminator().to(device)
    optimizer_G = torch.optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
    optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))
    scaler_G = GradScaler(device)
    scaler_D = GradScaler(device)
    scheduler_G = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_G, mode='min', factor=0.5, patience=10, verbose=True)  # Reduced patience
    scheduler_D = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_D, mode='min', factor=0.5, patience=10, verbose=True)  # Reduced patience

    start_epoch = 0
    best_g_loss = float('inf')
    g_losses, d_losses = [], []
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    checkpoint_path = os.path.join(checkpoint_dir, f"latest_checkpoint_{timestamp}.pth")

    print("Starting training from scratch.")

    start_time = time.time()

    for epoch in range(start_epoch, num_epochs):
        generator.train()
        discriminator.train()
        g_loss_total, d_loss_total = 0, 0

        for i, (real_images, cond) in enumerate(tqdm(train_loader, desc=f'Epoch {epoch+1}')):
            batch_size = real_images.size(0)
            real_images, cond = real_images.to(device), cond.to(device)

            # Clip inputs to avoid extreme values
            real_images = torch.clamp(real_images, -1.0, 1.0)
            cond = torch.clamp(cond, 0.0, 1.0)

            if i % 3 == 0:
                optimizer_D.zero_grad()
                with autocast(device_type=device):
                    noise = torch.randn(batch_size, noise_dim).to(device)
                    fake_images = generator(noise, cond)
                    fake_images = torch.clamp(fake_images, -1.0, 1.0)  # Clip fake images
                    real_validity, real_aux = discriminator(real_images, cond)
                    d_loss_real = F.binary_cross_entropy_with_logits(real_validity, torch.ones_like(real_validity) + 1e-8)  # Add epsilon
                    d_loss_aux = F.binary_cross_entropy_with_logits(real_aux, cond + 1e-8)  # Add epsilon
                    d_loss_real += 0.1 * d_loss_aux

                    fake_validity, fake_aux = discriminator(fake_images.detach(), cond)
                    d_loss_fake = F.binary_cross_entropy_with_logits(fake_validity, torch.zeros_like(fake_validity) + 1e-8)  # Add epsilon
                    d_loss_fake += 0.1 * F.binary_cross_entropy_with_logits(fake_aux, cond + 1e-8)  # Add epsilon

                    gradient_penalty = compute_gradient_penalty(discriminator, real_images, fake_images.detach(), cond, device)
                    d_loss = (d_loss_real + d_loss_fake) / 2 + 1.0 * gradient_penalty  # Reduced penalty weight

                scaler_D.scale(d_loss).backward()
                torch.nn.utils.clip_grad_norm_(discriminator.parameters(), max_norm=0.5)  # Reduced max_norm
                scaler_D.step(optimizer_D)
                scaler_D.update()
                d_loss_total += d_loss.item()

            optimizer_G.zero_grad()
            with autocast(device_type=device):
                noise = torch.randn(batch_size, noise_dim).to(device)
                fake_images = generator(noise, cond)
                fake_images = torch.clamp(fake_images, -1.0, 1.0)  # Clip fake images
                fake_validity, fake_aux = discriminator(fake_images, cond)
                g_loss_adv = F.binary_cross_entropy_with_logits(fake_validity, torch.ones_like(fake_validity) + 1e-8)  # Add epsilon
                g_loss_aux = F.binary_cross_entropy_with_logits(fake_aux, cond + 1e-8)  # Add epsilon
                g_loss = g_loss_adv + 0.1 * g_loss_aux
            scaler_G.scale(g_loss).backward()
            torch.nn.utils.clip_grad_norm_(generator.parameters(), max_norm=0.5)  # Reduced max_norm
            scaler_G.step(optimizer_G)
            scaler_G.update()
            g_loss_total += g_loss.item()

        g_loss_total /= len(train_loader)
        d_loss_total /= len(train_loader) // 3
        g_losses.append(g_loss_total)
        d_losses.append(d_loss_total)
        scheduler_G.step(g_loss_total)
        scheduler_D.step(d_loss_total)

        print(f'Epoch {epoch+1}:')
        print(f'Generator Loss: {g_loss_total:.4f} (Adv: {g_loss_adv.item():.4f}, Aux: {g_loss_aux.item():.4f})')
        print(f'Discriminator Loss: {d_loss_total:.4f} (Real: {d_loss_real.item():.4f}, Fake: {d_loss_fake.item():.4f}, GP: {gradient_penalty.item():.4f})')
        print(f'Learning Rate G: {optimizer_G.param_groups[0]["lr"]:.6f}, D: {optimizer_D.param_groups[0]["lr"]:.6f}')

        if g_loss_total < best_g_loss * 0.99 and not np.isnan(g_loss_total):
            best_g_loss = g_loss_total
            torch.save({
                'generator_state_dict': generator.state_dict(),
                'discriminator_state_dict': discriminator.state_dict(),
                'optimizer_G_state_dict': optimizer_G.state_dict(),
                'optimizer_D_state_dict': optimizer_D.state_dict(),
                'epoch': epoch,
                'best_g_loss': best_g_loss,
                'g_losses': g_losses,
                'd_losses': d_losses
            }, os.path.join(checkpoint_dir, f"best_gan_model_{timestamp}.pth"))
            print("Saved new best model.")

        if (epoch + 1) % checkpoint_interval == 0:
            torch.save({
                'generator_state_dict': generator.state_dict(),
                'discriminator_state_dict': discriminator.state_dict(),
                'optimizer_G_state_dict': optimizer_G.state_dict(),
                'optimizer_D_state_dict': optimizer_D.state_dict(),
                'epoch': epoch,
                'best_g_loss': best_g_loss,
                'g_losses': g_losses,
                'd_losses': d_losses
            }, checkpoint_path)
            print(f"Saved checkpoint at epoch {epoch+1}")

        if (epoch + 1) % 10 == 0:
            torch.cuda.empty_cache()
            gc.collect()
            generator.eval()
            with torch.no_grad():
                n_samples = 4
                cond = torch.zeros((n_samples, 40), device=device)
                cond[:, 20] = 1
                noise = torch.randn(n_samples, noise_dim).to(device)
                with autocast(device_type=device):
                    samples = generator(noise, cond)
                samples = (samples + 1) / 2
                img_grid = make_grid(samples.cpu())
                img_array = (img_grid.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
                plt.imshow(img_array)
                plt.title(f"Generated Samples (Condition: {train_dataset.attr_names[20]}) at Epoch {epoch+1}")
                plt.axis('off')
                plt.savefig(os.path.join(sample_dir, f"samples_epoch_{epoch+1}_{timestamp}.png"))
                plt.show()

    end_time = time.time()
    total_training_time = end_time - start_time
    print(f"Total training time: {total_training_time:.2f} seconds ({total_training_time / 3600:.2f} hours)")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(g_losses, label='Generator Loss')
    plt.plot(d_losses, label='Discriminator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Losses')
    plt.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(checkpoint_dir, f"loss_plot_{timestamp}.png"))
    plt.show()

if __name__ == '__main__':
    train_and_evaluate()
```
