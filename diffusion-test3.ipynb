```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.amp import GradScaler, autocast
from torchvision import transforms, models
from torchvision.utils import make_grid
from PIL import Image
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
import gc

# Dataset class with improved filename matching and debugging
class CelebAHQDataset(Dataset):
    def __init__(self, data_dir, attr_file, transform=None, split='train'):
        self.data_dir = data_dir
        print(f"Data directory: {self.data_dir}")
        self.transform = transform

        print(f"Attribute file: {attr_file}")
        if attr_file.endswith('.xlsx'):
            self.attributes_df = pd.read_excel(attr_file, engine='openpyxl')
        else:
            self.attributes_df = pd.read_csv(attr_file)

        self.attr_names = self.attributes_df.columns[1:].tolist()
        self.attr_filenames = [os.path.basename(str(f)).strip().lower() for f in self.attributes_df.iloc[:, 0].tolist()]
        img_files = [f.strip().lower() for f in os.listdir(self.data_dir) if f.lower().endswith(('.jpg', '.png'))]

        print("\U0001F4BE First 5 filenames from CSV:", self.attr_filenames[:5])
        print("\U0001F5BC First 5 filenames from folder:", img_files[:5])

        filename_to_idx = {name: idx for idx, name in enumerate(self.attr_filenames)}
        self.matched_files, self.matched_indices = [], []

        for img_file in img_files:
            if img_file in filename_to_idx:
                self.matched_files.append(img_file)
                self.matched_indices.append(filename_to_idx[img_file])

        self.attributes = self.attributes_df.iloc[self.matched_indices, 1:]
        split_idx = int(0.8 * len(self.matched_files))
        if split == 'train':
            self.matched_files = self.matched_files[:split_idx]
            self.attributes = self.attributes.iloc[:split_idx]
        else:
            self.matched_files = self.matched_files[split_idx:]
            self.attributes = self.attributes.iloc[split_idx:]

        # Debug: Print sample attribute values
        print("Sample attributes (first 5 samples):", self.attributes.head().values)
        print(f"\u2705 Matched {len(self.matched_files)} images out of {len(img_files)}")

    def __len__(self):
        return len(self.matched_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.data_dir, self.matched_files[idx])
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        attrs = (self.attributes.iloc[idx].values + 1) // 2  # Normalize to [0, 1]
        return image, torch.tensor(attrs, dtype=torch.float32)

class FiLM(nn.Module):
    def __init__(self, channels, cond_dim):
        super().__init__()
        self.gamma = nn.Linear(cond_dim, channels)
        self.beta = nn.Linear(cond_dim, channels)
        nn.init.xavier_uniform_(self.gamma.weight)
        nn.init.zeros_(self.gamma.bias)
        nn.init.xavier_uniform_(self.beta.weight)
        nn.init.zeros_(self.beta.bias)

    def forward(self, x, cond):
        gamma = self.gamma(cond).view(x.size(0), x.size(1), 1, 1)
        beta = self.beta(cond).view(x.size(0), x.size(1), 1, 1)
        return x * gamma + beta

class FiLMBlock(nn.Module):
    def __init__(self, in_ch, out_ch, cond_dim):
        super().__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, 3, padding=1)
        self.act1 = nn.SiLU()
        self.film = FiLM(out_ch, cond_dim)
        self.norm = nn.GroupNorm(32, out_ch)
        self.conv2 = nn.Conv2d(out_ch, out_ch, 3, padding=1)
        self.act2 = nn.SiLU()
        nn.init.xavier_uniform_(self.conv1.weight)
        nn.init.zeros_(self.conv1.bias)
        nn.init.xavier_uniform_(self.conv2.weight)
        nn.init.zeros_(self.conv2.bias)

    def forward(self, x, cond):
        x = self.conv1(x)
        x = self.act1(x)
        x = self.film(x, cond)
        x = self.norm(x)
        x = self.conv2(x)
        x = self.act2(x)
        return x

class ConditionalUNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, cond_dim=40, time_dim=128):
        super().__init__()
        self.time_dim = time_dim
        self.cond_dim = cond_dim
        self.time_embed = nn.Sequential(
            nn.Linear(time_dim, time_dim * 4),
            nn.SiLU(),
            nn.Linear(time_dim * 4, time_dim),
        )
        self.init_conv = nn.Conv2d(in_channels, 64, 3, padding=1)
        self.down1 = FiLMBlock(64, 128, cond_dim + time_dim)
        self.down2 = FiLMBlock(128, 256, cond_dim + time_dim)
        self.down3 = FiLMBlock(256, 512, cond_dim + time_dim)
        self.bottleneck = nn.Sequential(
            nn.Conv2d(512, 512, 3, padding=1),
            nn.SiLU(),
            nn.Conv2d(512, 512, 3, padding=1),
        )
        self.up1 = FiLMBlock(512 + 256, 256, cond_dim + time_dim)
        self.up2 = FiLMBlock(256 + 128, 128, cond_dim + time_dim)
        self.up3 = FiLMBlock(128 + 64, 64, cond_dim + time_dim)
        self.out_conv = nn.Sequential(
            nn.Conv2d(64, out_channels, 1),
            nn.Tanh()
        )
        self.classifier = nn.Sequential(
            nn.Conv2d(512, 256, 3, padding=1),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(256, cond_dim)
        )
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, x, t, cond):
        t_emb = self.get_time_embedding(t)
        cond_t = torch.cat([cond, t_emb], dim=1)
        x1 = self.init_conv(x)
        x2 = self.down1(F.max_pool2d(x1, 2), cond_t)
        x3 = self.down2(F.max_pool2d(x2, 2), cond_t)
        x4 = self.down3(F.max_pool2d(x3, 2), cond_t)
        x = self.bottleneck(x4)
        attr_pred = self.classifier(x)
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=False)
        x = self.up1(torch.cat([x, x3], dim=1), cond_t)
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=False)
        x = self.up2(torch.cat([x, x2], dim=1), cond_t)
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=False)
        x = self.up3(torch.cat([x, x1], dim=1), cond_t)
        return self.out_conv(x), attr_pred

    def get_time_embedding(self, t):
        half_dim = self.time_dim // 2
        emb = torch.log(torch.tensor(10000.0)) / (half_dim - 1)
        emb = torch.exp(torch.arange(half_dim, device=t.device) * -emb)
        emb = t[:, None] * emb[None, :]
        emb = torch.cat([torch.sin(emb), torch.cos(emb)], dim=1)
        return self.time_embed(emb)

class PerceptualLoss(nn.Module):
    def __init__(self):
        super().__init__()
        vgg = models.vgg16(pretrained=True).features[:16]
        self.vgg_layers = vgg.eval()
        for param in self.vgg_layers.parameters():
            param.requires_grad = False

    def forward(self, x, y):
        x_vgg = self.vgg_layers((x + 1) / 2)  # Denormalize from [-1, 1] to [0, 1]
        y_vgg = self.vgg_layers((y + 1) / 2)
        return F.mse_loss(x_vgg, y_vgg)

class DiffusionScheduler:
    def __init__(self, num_timesteps=1000, beta_start=0.0001, beta_end=0.02):
        self.num_timesteps = num_timesteps
        self.beta_start = beta_start
        self.beta_end = beta_end
        self.betas = self.cosine_schedule().to(device='cuda' if torch.cuda.is_available() else 'cpu')
        self.alphas = 1.0 - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)
        self.sqrt_alphas_cumprod = torch.sqrt(self.alphas_cumprod)
        self.sqrt_one_minus_alphas_cumprod = torch.sqrt(1.0 - self.alphas_cumprod)

    def cosine_schedule(self):
        steps = self.num_timesteps + 1
        x = np.linspace(0, self.num_timesteps, steps)
        alphas_cumprod = np.cos(0.5 * np.pi * x / steps) ** 2
        alphas_cumprod = alphas_cumprod / alphas_cumprod[0]
        betas = 1 - alphas_cumprod[1:] / alphas_cumprod[:-1]
        return torch.tensor(np.clip(betas, a_min=self.beta_start, a_max=self.beta_end * 0.999), dtype=torch.float32)

    def add_noise(self, x_0, t, noise):
        sqrt_alpha_prod = self.sqrt_alphas_cumprod[t].view(-1, 1, 1, 1)
        sqrt_one_minus_alpha_prod = self.sqrt_one_minus_alphas_cumprod[t].view(-1, 1, 1, 1)
        return sqrt_alpha_prod * x_0 + sqrt_one_minus_alpha_prod * noise

class GaussianDiffusion:
    def __init__(self, scheduler, model):
        self.scheduler = scheduler
        self.model = model
        self.perceptual_loss = PerceptualLoss().to(device='cuda' if torch.cuda.is_available() else 'cpu')

    def sample_timesteps(self, batch_size, device):
        return torch.randint(0, self.scheduler.num_timesteps, (batch_size,), device=device)

    def training_step(self, x_0, cond, device):
        with autocast(device_type=device):
            t = self.sample_timesteps(x_0.size(0), device)
            noise = torch.randn_like(x_0, device=device)
            x_t = self.scheduler.add_noise(x_0, t, noise)
            generated_img, attr_pred = self.model(x_t, t, cond)
            if torch.isnan(generated_img).any() or torch.isnan(attr_pred).any():
                print(f"Warning: NaN detected in generated_img or attr_pred at timestep {t[0].item()}")
            noise_loss = F.mse_loss(generated_img, noise)
            attr_loss = F.binary_cross_entropy_with_logits(attr_pred, cond)
            if torch.isnan(noise_loss) or torch.isnan(attr_loss):
                print(f"Warning: NaN detected in loss calculation")
            attr_loss_weight = 2.0  # Reduced from 5.0 to prevent domination
            loss = noise_loss + attr_loss * attr_loss_weight
            if self.perceptual_loss is not None:
                clean_img_pred, _ = self.model(torch.zeros_like(x_0), torch.zeros_like(t), cond)
                perceptual = self.perceptual_loss(clean_img_pred, x_0)
                loss += perceptual * 0.1
            return loss, attr_loss * attr_loss_weight, attr_pred

    def sample(self, shape, cond, device, num_steps=1000):
        x = torch.randn(shape, device=device)
        for t in reversed(range(0, num_steps)):
            t_batch = torch.full((shape[0],), t, device=device, dtype=torch.long)
            with autocast(device_type=device):
                pred_noise, _ = self.model(x, t_batch, cond)
                alpha = self.scheduler.alphas[t]
                alpha_cumprod = self.scheduler.alphas_cumprod[t]
                beta = self.scheduler.betas[t]
                x = (x - (1 - alpha) / torch.sqrt(1 - alpha_cumprod) * pred_noise) / torch.sqrt(alpha)
                if t > 0:
                    x += torch.sqrt(beta) * torch.randn_like(x)
        return torch.clamp(x, -1, 1)

def train_and_evaluate():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    data_dir = r"C:\Users\USER\FaceGeneration_CelebAHQ\img_align_celeba\img_align_celeba"
    attr_file = r"C:\Users\USER\FaceGeneration_CelebAHQ\img_align_celeba\list_attr_celeba.csv"
    batch_size = 64  # Changed from 128 to 64
    num_epochs = 300
    lr = 5e-5  # Reduced from 1e-4 to prevent rapid convergence
    num_timesteps = 1000
    checkpoint_interval = 10  # Save checkpoint every 10 epochs
    checkpoint_dir = "checkpoints"
    os.makedirs(checkpoint_dir, exist_ok=True)

    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    ])

    train_dataset = CelebAHQDataset(data_dir, attr_file, transform, split='train')
    val_dataset = CelebAHQDataset(data_dir, attr_file, transform, split='val')
    print(f"Train dataset size: {len(train_dataset)}, Val dataset size: {len(val_dataset)}")
    
    train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
    for images, attrs in train_loader:
        plt.imshow(make_grid(images).permute(1, 2, 0).cpu().numpy())
        plt.title([train_dataset.attr_names[i] for i in attrs[0].nonzero()[0]])
        plt.show()
        break

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    # Memory check
    try:
        dummy_input = torch.randn(1, 3, 128, 128).to(device)
        model = ConditionalUNet().to(device)
        _ = model(dummy_input, torch.tensor([0]).to(device), torch.zeros(1, 40).to(device))
        print("Memory check passed. Batch size 64 should be fine.")
    except RuntimeError as e:
        print(f"Memory check failed: {e}. Consider reducing batch_size further.")
        return

    model = ConditionalUNet().to(device)
    scheduler = DiffusionScheduler(num_timesteps=num_timesteps)
    diffusion = GaussianDiffusion(scheduler, model)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    scaler = GradScaler(device)
    lr_scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    # Load checkpoint if exists
    start_epoch = 0
    best_val_loss = float('inf')
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    checkpoint_path = os.path.join(checkpoint_dir, "latest_checkpoint.pth")

    if os.path.exists(checkpoint_path):
        checkpoint = torch.load(checkpoint_path, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        lr_scheduler.load_state_dict(checkpoint['lr_scheduler_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        best_val_loss = checkpoint['best_val_loss']
        train_losses = checkpoint['train_losses']
        val_losses = checkpoint['val_losses']
        train_accs = checkpoint['train_accs']
        val_accs = checkpoint['val_accs']
        print(f"Resuming training from epoch {start_epoch} with best_val_loss {best_val_loss:.4f}")
    else:
        print("No checkpoint found. Starting training from scratch.")

    for epoch in range(start_epoch, num_epochs):
        model.train()
        train_noise_loss, train_attr_loss, train_correct, train_total = 0, 0, 0, 0
        for images, attrs in tqdm(train_loader, desc=f'Epoch {epoch+1}'):
            images, attrs = images.to(device), attrs.to(device)
            optimizer.zero_grad()
            with autocast(device_type=device):
                loss, attr_loss, attr_pred = diffusion.training_step(images, attrs, device)
            scaler.scale(loss).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
            train_noise_loss += (loss.item() - attr_loss.item())
            train_attr_loss += attr_loss.item()
            preds = (torch.sigmoid(attr_pred) > 0.5).float()
            train_correct += (preds == attrs).sum().item()
            train_total += attrs.numel()

        train_noise_loss /= len(train_loader)
        train_attr_loss /= len(train_loader)
        train_acc = train_correct / train_total
        train_losses.append(train_noise_loss + train_attr_loss)
        train_accs.append(train_acc)

        model.eval()
        val_noise_loss, val_attr_loss, val_correct, val_total = 0, 0, 0, 0
        val_preds, val_labels = [], []
        with torch.no_grad():
            for images, attrs in val_loader:
                images, attrs = images.to(device), attrs.to(device)
                with autocast(device_type=device):
                    loss, attr_loss, attr_pred = diffusion.training_step(images, attrs, device)
                val_noise_loss += (loss.item() - attr_loss.item())
                val_attr_loss += attr_loss.item()
                preds = (torch.sigmoid(attr_pred) > 0.5).float()
                val_correct += (preds == attrs).sum().item()
                val_total += attrs.numel()
                val_preds.append(torch.sigmoid(attr_pred).cpu().numpy())
                val_labels.append(attrs.cpu().numpy())

        val_noise_loss /= len(val_loader)
        val_attr_loss /= len(val_loader)
        val_acc = val_correct / val_total
        val_losses.append(val_noise_loss + val_attr_loss)
        val_accs.append(val_acc)

        lr_scheduler.step()

        print(f'Epoch {epoch+1}:')
        print(f'Train Loss: {train_noise_loss + train_attr_loss:.4f} (Noise: {train_noise_loss:.4f}, Attr: {train_attr_loss:.4f})')
        print(f'Train Acc: {train_acc:.4f}')
        print(f'Val Loss: {val_noise_loss + val_attr_loss:.4f} (Noise: {val_noise_loss:.4f}, Attr: {val_attr_loss:.4f})')
        print(f'Val Acc: {val_acc:.4f}')
        print(f'Learning Rate: {lr_scheduler.get_last_lr()[0]:.6f}')

        total_val_loss = val_noise_loss + val_attr_loss
        if total_val_loss < best_val_loss * 0.99:
            best_val_loss = total_val_loss
            torch.save(model.state_dict(), "best_model.pth")
            print("Saved new best model.")

        # Save checkpoint
        if (epoch + 1) % checkpoint_interval == 0:
            checkpoint = {
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'lr_scheduler_state_dict': lr_scheduler.state_dict(),
                'best_val_loss': best_val_loss,
                'train_losses': train_losses,
                'val_losses': val_losses,
                'train_accs': train_accs,
                'val_accs': val_accs
            }
            torch.save(checkpoint, checkpoint_path)
            print(f"Saved checkpoint at epoch {epoch+1}")

        if (epoch + 1) % 10 == 0:
            torch.cuda.empty_cache()
            gc.collect()
            model.eval()
            with torch.no_grad():
                n_samples = 2
                cond = torch.zeros((n_samples, 40), device=device)
                cond[:, 20] = 1  # Condition on the 21st attribute (e.g., Male)
                with autocast(device_type=device):
                    samples = diffusion.sample((n_samples, 3, 128, 128), cond, device)
                samples = (samples * 0.5 + 0.5).clamp(0, 1)
                plt.imshow(make_grid(samples).permute(1, 2, 0).cpu().numpy())
                plt.title(f"Generated Samples (Condition: {train_dataset.attr_names[20]}) at Epoch {epoch+1}")
                plt.axis('off')
                plt.show()

                # Reconstruction visualization
                x, cond = next(iter(val_loader))
                x, cond = x.to(device), cond.to(device)
                x_t = scheduler.add_noise(x, torch.zeros(x.size(0), dtype=torch.long, device=device), torch.randn_like(x))
                x_hat, _ = model(x_t, torch.zeros(x.size(0), dtype=torch.long, device=device), cond)
                x_hat = (x_hat * 0.5 + 0.5).clamp(0, 1)
                x = (x * 0.5 + 0.5).clamp(0, 1)
                grid = make_grid(torch.cat([x, x_hat]), nrow=x.size(0))
                plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
                plt.title("Original (top) vs Reconstructed (bottom)")
                plt.show()

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    val_preds = np.concatenate(val_preds)
    val_labels = np.concatenate(val_labels)
    for i, attr_name in enumerate(train_dataset.attr_names[:3]):
        fpr, tpr, _ = roc_curve(val_labels[:, i], val_preds[:, i])
        auc = roc_auc_score(val_labels[:, i], val_preds[:, i])
        plt.plot(fpr, tpr, label=f'{attr_name} (AUC={auc:.2f})')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves')
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    train_and_evaluate()
```
